**Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Loading the csv**

```python
data = pd.read_csv('/content/sample_data/Airbnb NYC 2019.csv')
```

```python
# Checking for null values
data.isnull().sum()
```

```python
# Dropping rows with null values
data.dropna(subset =['name'],inplace = True)
```

```python
data['name'].isnull().sum()
```

```python
#dropping columns with null host name
data.dropna(subset=['host_name'], inplace=True)
```

```python
data['host_name'].isnull().sum()
```

```python
# Filling the null values with zero
data['reviews_per_month']=data['reviews_per_month'].fillna(0, inplace=True)
data['last_review']=data['last_review'].fillna(0, inplace=True)
```

```python
# Converting the last review column data type
data['last_review'] = pd.to_datetime(data['last_review'], errors='coerce')
```

```python
data['last_review'].dtype
```

```python
# Columns with potential outliers for visual inspection
outlier_columns = ['price', 'minimum_nights', 'number_of_reviews',]

# Set up the plotting area
plt.figure(figsize=(16, 10))

# Plot boxplots to detect outliers
for i, column in enumerate(outlier_columns, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x=data[column].dropna())
    plt.title(f'Boxplot of {column}')

plt.tight_layout()
plt.show()
```

```python
# Calculate the average price per room type
average_price_per_room_type = data.groupby('room_type')['price'].mean().reset_index()

# Display the result
average_price_per_room_type
```

**Visualization**

```python
# Create the violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='room_type', y='price', data=data, palette='Set3')

# Customize plot appearance
plt.title('Distribution of Prices by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
#Calculate total counts of each room type
room_type_counts = data['room_type'].value_counts()

# Step 2: Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(room_type_counts, labels=room_type_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Total Percentage of Room Types')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is a circle.
plt.show()
```

```python
# Create a bar chart
avg_price_per_room_type = data.groupby('room_type')['price'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='room_type', y='price', data=avg_price_per_room_type, palette='Set3')

# Add titles and labels
plt.title('Average Price by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.legend(['Average Price'], loc='upper right')  # Add a legend
# Add values on top of each bar
for index, row in avg_price_per_room_type.iterrows():
    plt.text(index, row['price'] + 5, round(row['price'], 2), ha='center', va='bottom')

plt.show()
```

```python
# Create a bar plot
# Define price bins
bins = [0, 100, 200, 300, 400, 500, 1000]
labels = ['0-100', '100-200', '200-300', '300-400', '400-500', '500+']
data['price_range'] = pd.cut(data['price'], bins=bins, labels=labels)

# Create a count plot (bar chart) with hue for price range
plt.figure(figsize=(12, 6))
count_plot = sns.countplot(x='room_type', hue='price_range', data=data, palette='Set3')

# Add values on top of each bar
for p in count_plot.patches:
    count_plot.annotate(
        f'{int(p.get_height())}',  # Display as an integer
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center', va='baseline', fontsize=9, color='black', xytext=(0, 5),
        textcoords='offset points'
    )

# Add titles and labels
plt.title('Count of Listings by Room Type and Price Range')
plt.xlabel('Room Type')
plt.ylabel('Count of Listings')
plt.xticks(rotation=45)
plt.legend(title='Price Range')

plt.show()
```

```python
# Calculate the average price per neighborhood group
avg_price_by_neighborhood = data.groupby('neighbourhood_group')['price'].mean().reset_index()

# Create a bar chart
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x='neighbourhood_group', y='price', data=avg_price_by_neighborhood, palette='Set2')

# Add values on top of each bar
for index, row in avg_price_by_neighborhood.iterrows():
    plt.text(index, row['price'] + 5, round(row['price'], 2), ha='center', va='bottom')

# Add titles and labels
plt.title('Average Price by Neighborhood Group')
plt.xlabel('neighbourhood_group')
plt.ylabel('Average Price')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

plt.show()
```

```python
#Group by neighborhood group and room type, and count occurrences
room_type_counts = data.groupby(['neighbourhood_group', 'room_type']).size().unstack()

# Step 2: Calculate proportions
room_type_proportions = room_type_counts.div(room_type_counts.sum(axis=1), axis=0)

# Step 3: Plotting the stacked bar chart
ax = room_type_proportions.plot(kind='bar', stacked=True, figsize=(10, 6))

# Adding the numbers on the bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height() * room_type_counts.sum(axis=1).sum())}',
                (p.get_x() + p.get_width() / 2., p.get_y() + p.get_height()),
                ha='center', va='bottom')

plt.title('Room Type Proportion by Neighborhood Group')
plt.xlabel('Neighborhood Group')
plt.ylabel('Proportion of Room Types')
plt.legend(title='Room Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
